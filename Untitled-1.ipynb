```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering , k_means
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn import metrics
from sklearn.metrics import silhouette_score

```

```python
df=pd.read_csv("data.csv")

x=df.iloc[:, [3,4]].values #or df=df.to_numpy() same result 
print (df)
colors=np.random.rand(200)
plt.scatter(x[: , 0], x[:, 1], c=colors, marker='o', s=120 )
plt.show()
```

```python

silhouette_full_date =[]
silhouette_choosed_date =[]
intra=[]
k=range(2,10)

for i in k:
 cluster1=k_means(n_clusters=i)
 lable=cluster1.fit_predict(x)
 intra.append(cluster1.inertia_)
    
    
 silhouette_full_date.append(metrics.silhouette_score(df , lable , metric='euclidean'))
 silhouette_choosed_date.append(metrics.silhouette_score(x , lable , metric='euclidean'))
 print (f'the silhouette for k {i} the whole date is {silhouette_full_date} ')
 print (f'the silhouette for k {i} the choosed date is {silhouette_choosed_date} ')
  
 
  
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k, intra, marker='o', color='blue')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')

plt.subplot(1, 2, 2)
plt.plot(k, silhouette_full_date, marker='x', color='green', label='Full Dataset')
plt.plot(k, silhouette_choosed_date, marker='o', color='orange', label='Chosen Dataset')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k')
plt.legend()
plt.tight_layout()
plt.show()
  
  
 
    
    
    
    
    
    
    
```

```python
color=10*["g","r","c","m","b"]
for v in ['ward', 'complete', 'average', 'single']:
    model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage=v)    
    model.fit(x)
    cluster_lable=model.labels_
    cm=[color[i] for i in cluster_lable.astype(int)]
    plt.title(str.format('Clustering for linkage: %s" %v'))
    plt.scatter(x[: , 0] , x[ : , 1] , marker='x' , c=cm)
    plt.show()
```
