```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('penguins.csv')
df.describe
df = df.dropna(axis=0) #remove of empty row if axis = 1 remove of coulum 
df = df[['bill_length_mm', 'flipper_length_mm']]
#print(df)
df.head()
```

```python

colors =np.random.rand(333)
#area = (30* np.random.rand(333))**2
plt.scatter(df['bill_length_mm'], df['flipper_length_mm'], c=colors, marker='o', s=120 )
plt.xlabel("bill_length_mm")
plt.ylabel("flipper_length_mm")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics

intertias=[]
silk=[]
k=range(2,10)
for i in k:
    
 cluster=KMeans(n_clusters=i)
 labels=cluster.fit_predict(df)
 intertias.append(cluster.inertia_)
 silk.append(metrics.silhouette_score(df , labels , metric='euclidean' ))

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(k, intertias , c='c')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")

plt.subplot(1,2,2)
plt.plot(k, silk, marker='o', color='r')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different k")
plt.tight_layout()
plt.show()

plt.show()
```
